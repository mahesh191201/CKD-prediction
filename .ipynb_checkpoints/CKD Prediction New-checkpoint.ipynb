```python
df.drop('id',axis=1,inplace=True)
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

col = ['rbc', 'pcc', 'pc', 'ba', 'htn', 'dm', 'cad', 'pe', 'ane']
encoder = LabelEncoder()
for col in col:
    df[col] = encoder.fit_transform(df[col])
```

```python
df[['appet', 'classification']] = df[['appet', 'classification']].replace(to_replace={'good':'1', 'ckd':'1', 'notckd':'0', 'poor':'0'})
```

```python
count=df["classification"].value_counts()
print("Percentage of patience with heart problems: "+str(round(count[1]*100/303))+"%")
print("Percentage of patience without heart problems: "+str(round(count[0]*100/303))+"%")


print("No of people with CK disease: ", len(df[df["classification"]==0]))
print("No of people without CK disease: ", len(df[df["classification"]==1]))
```

```python
from sklearn.model_selection import train_test_split
data = df.drop("classification", axis=1)
target = df["classification"]
X_train, X_test, Y_train, Y_test = train_test_split(data, target, test_size=0.30)
```

```python
Y_test.head(10)
```

```python
X_train.shape, Y_train.shape, X_test.shape, Y_test.shape
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.linear_model import LogisticRegression
LR = LogisticRegression()
LR.fit(X_train, Y_train)
```

```python
LrPred=LR.predict(X_test)
LrScore= round(accuracy_score(LrPred, Y_test)*100)
print("The score of Logistic Regression model is: "+str(LrScore)+"%")
```

```python
from sklearn.naive_bayes import GaussianNB
GNB = GaussianNB()
GNB.fit(X_train, Y_train)
GNBPred=GNB.predict(X_test)
GNBScore= round(accuracy_score(GNBPred, Y_test)*100)
print("The score of Logistic Regression model is: "+str(GNBScore)+"%")

```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN=KNeighborsClassifier(n_neighbors=7)
KNN.fit(X_train, Y_train)
KNN_pred=KNN.predict(X_test)

KNN_score=round(accuracy_score(KNN_pred, Y_test)*100)
print("The accuracy of KNN is: "+str(KNN_score)+"%")
```

```python
from sklearn.tree import DecisionTreeClassifier

max_accuracy=0;

for x in range(2000):
    model=DecisionTreeClassifier(random_state=x)
    model.fit(X_train, Y_train)
    DT_pred=model.predict(X_test)
    curr_accuracy=round(accuracy_score(DT_pred, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
model=DecisionTreeClassifier(random_state=best_x)
model.fit(X_train, Y_train)
DT_pred=model.predict(X_test)

DT_score=round(accuracy_score(DT_pred, Y_test)*100)
print("The accuracy of Decision Tree is: "+str(DT_score)+"%")
```

```python
from sklearn.ensemble import RandomForestClassifier

max_accuracy=0;

for x in range(200):
    model=RandomForestClassifier(random_state=x)
    model.fit(X_train, Y_train)
    Y_pred_rf=model.predict(X_test)
    curr_accuracy=round(accuracy_score(Y_pred_rf, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
rf=DecisionTreeClassifier(random_state=best_x)
rf.fit(X_train, Y_train)
Y_pred_rf=rf.predict(X_test)

RF_score=round(accuracy_score(Y_pred_rf, Y_test)*100)
print("The accuracy of Random Forest is: "+str(RF_score)+"%")
```

```python
import xgboost as xgb

xgb_model=xgb.XGBClassifier(objective="binary:logistic", random_state=42)
xgb_model.fit(X_train, Y_train)

Y_pred_xgb=xgb_model.predict(X_test)

XGB_score=round(accuracy_score(Y_pred_xgb, Y_test)*100)
print("The accuracy of XGBoost is: "+str(XGB_score)+"%")
```

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential()
model.add(Dense(12, activation='relu', input_dim=24))
model.add(Dense(1, activation='sigmoid'))

model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, Y_train, epochs=300)
```

```python
Y_pred_nn = model.predict(X_test)

rounded = [round(x[0]) for x in Y_pred_nn]

Y_pred_nn = rounded
NN_score = round(accuracy_score(Y_pred_nn,Y_test)*100)

print("The accuracy score achieved using Neural Network is: "+str(NN_score)+"%")
```

```python
from sklearn.ensemble import AdaBoostClassifier


max_accuracy=0;

for x in range(200):
    model=AdaBoostClassifier(random_state=x)
    model.fit(X_train, Y_train)
    Y_pred_rf=model.predict(X_test)
    curr_accuracy=round(accuracy_score(Y_pred_rf, Y_test)*100)
    if(curr_accuracy>max_accuracy):
        max_accuracy=curr_accuracy
        best_x=x
        
rf=DecisionTreeClassifier(random_state=best_x)
rf.fit(X_train, Y_train)
Y_pred_rf=rf.predict(X_test)

RF_score=round(accuracy_score(Y_pred_rf, Y_test)*100)
print("The accuracy of Adaboost is: "+str(RF_score)+"%")
```

```python
from sklearn.ensemble import AdaBoostClassifier
import pickle
fileName= "AdaBoost.pkl"

AC= AdaBoostClassifier(n_estimators=500,  learning_rate=0.1)
AC=AC.fit(X_train, Y_train)
pickle.dump(AC, open(fileName, "wb"))

AC_pred=AC.predict(X_test)

AC_score=round(accuracy_score(AC_pred, Y_test)*100)
print("The accuracy of AC_pred is: "+str(AC_score)+"%")
```
